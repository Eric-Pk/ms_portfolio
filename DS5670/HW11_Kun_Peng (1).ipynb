### Train the Model

```python
from gensim.models import Word2Vec

# Train Word2Vec model
model = Word2Vec(sentences=preprocessed_sentences, # The corpus
                 vector_size=300, # What size the word vectors should be (more is usually better), 300 is usually default
                 window=5, # The context size for skip-gram
                 min_count=4, # Ignore words with fequency less than this
                 workers=4, # How many worker threads to employ (for multicore machines)
                 sg=1) # 1 for skip-gram, otherwise CBOW

# Save the model for future use
model.save("word2vec.model") # you would need to call it in. We won't do that today.
```

The `model.wv` object contains the mapping between the words and the embeddings. This is the object you will use for analysis.

```python
model.wv['home'].shape
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
# Reduce dimensionality of word vectors for visualization
word_vectors = model.wv[model.wv.index_to_key]  # Get the word vectors
pca = PCA(n_components=2)  # Initialize PCA
result = pca.fit_transform(word_vectors)  # Apply PCA to word vectors and transform the word vectors

# Plot the words in a 2D space
plt.figure(figsize=(10, 8))
plt.scatter(result[:, 0], result[:, 1]) #scatterplot for first and second component

# Annotate words in the plot
words = list(model.wv.index_to_key) #get words from vocab
for i, word in enumerate(words): #loop through words
    plt.annotate(word, xy=(result[i, 0], result[i, 1]), fontsize=12) #annotate them

plt.title("Word Embeddings Visualization")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.grid()
plt.show()
```

### Word2Vec Model Methods
- The Word2Vec model from Gensim comes with a number of useful methods that you can use out-of-the-box to investigate the similarity between words.

- You can play around with these to develop your intuitive understanding of how word vectors work!

#### You can use the `model.similarity()` method to get the similarity between two word vectors.
- Gensim uses cosine similarity (normalized dot product) of the two vectors.

```python
word1 = 'cat'
word2 = 'dog'

try: #good for error handling in case word not available.
    similarity = word2vec_model.similarity(word1, word2)
    print(f"Similarity between '{word1}' and '{word2}': {similarity:.4f}")
except KeyError as e:
    print(f"KeyError: {e}")
```

#### You can use the `model.most_similar(word, topn=num)` method to find the topn most similar words

```python
word = 'cat'

try:
    similar_words = word2vec_model.most_similar(word, topn=10)
    print(f"Most similar words to '{word}':", similar_words)
except KeyError as e:
    print(f"KeyError: {e}")
```

#### You can use the `model.most_similar(positive=list, negative=list, topn=num)` method to do analogy tasks.
- All we are doing here is getting the sum of vectors
- `positive` is a list of vectors we are adding
- `negative` is a list of vectors we are subtracting
- Finally, we take the most similar word vector to the sum.

```python
import pandas as pd
import numpy as np
import gensim.downloader as api
from sklearn.metrics.pairwise import cosine_similarity
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize

nltk.download('punkt')

# Load dataset from Google Drive (adjust your path accordingly)
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DS_5780_spring_25/writing_data_w_meta_data.csv")

# Load the Word2Vec pre-trained model
word2vec_model = api.load("word2vec-google-news-300")

# Compute average word similarity in an essay
def average_word_similarity(text):
    words = [w for w in word_tokenize(str(text).lower()) if w in word2vec_model]
    if len(words) < 2:
        return np.nan
    word_vectors = np.array([word2vec_model[w] for w in words])
    similarities = cosine_similarity(word_vectors)
    return np.mean(similarities[np.triu_indices(len(words), k=1)])  # Upper triangle only

# Compute average sentence similarity
def average_sentence_similarity(text):
    sentences = sent_tokenize(str(text))
    sentence_vectors = []

    for sent in sentences:
        words = [w for w in word_tokenize(sent.lower()) if w in word2vec_model]
        if words:
            sent_vec = np.mean([word2vec_model[w] for w in words], axis=0)  # Sentence vector is mean of word vectors
            sentence_vectors.append(sent_vec)

    if len(sentence_vectors) < 2:
        return np.nan

    similarities = cosine_similarity(sentence_vectors)
    return np.mean(similarities[np.triu_indices(len(sentence_vectors), k=1)])

# Compute adjacent sentence similarity
def adjacent_sentence_similarity(text):
    sentences = sent_tokenize(str(text))
    sentence_vectors = []

    for sent in sentences:
        words = [w for w in word_tokenize(sent.lower()) if w in word2vec_model]
        if words:
            sent_vec = np.mean([word2vec_model[w] for w in words], axis=0)
            sentence_vectors.append(sent_vec)

    if len(sentence_vectors) < 2:
        return np.nan

    similarities = [cosine_similarity([sentence_vectors[i]], [sentence_vectors[i+1]])[0][0]
                    for i in range(len(sentence_vectors)-1)]

    return np.mean(similarities)

# Apply the functions to compute similarity metrics
df["avg_word_similarity"] = df["text"].apply(average_word_similarity)
df["avg_sentence_similarity"] = df["text"].apply(average_sentence_similarity)
df["adjacent_sentence_similarity"] = df["text"].apply(adjacent_sentence_similarity)

# Correlation Analysis
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
sns.heatmap(df[['avg_word_similarity', 'avg_sentence_similarity', 'adjacent_sentence_similarity', 'Holistic', 'Unity']].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# Linear Regression Analysis
import statsmodels.api as sm

X = df[['avg_word_similarity', 'avg_sentence_similarity', 'adjacent_sentence_similarity']]
y = df['Holistic']  # Target variable: overall writing proficiency
X = sm.add_constant(X)  # Add constant term
model = sm.OLS(y, X, missing='drop').fit()

print(model.summary())  # Display regression results
```

### **Key Findings:**
**Overall Model Significance**:
   - The model is statistically significant (**F-statistic = 9.561, p < 0.0001**).
   - However, it explains **only 4.8% of the variance** in Holistic scores (**R = 0.048**), meaning other factors influence the writing score more than just these cohesion metrics.

**Coefficient Analysis**:
   - **`avg_word_similarity` (p < 0.001, negative effect)**  
      Higher word similarity **lowers** holistic scores.  
     **Interpretation**: If a student's words are too similar, it may indicate redundancy or lack of lexical diversity.
   - **`avg_sentence_similarity` (p = 0.189, not significant)**  
      No strong relationship with holistic scores.
   - **`adjacent_sentence_similarity` (p = 0.318, not significant)**  
      No clear effect on holistic scores.
